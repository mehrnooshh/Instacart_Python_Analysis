# 03 Data wrangling

## 3-1. Data wrangling for the examples that we walked through in the course

### 3-1-1. Deleting a column from a dataframe

```python
# Dropping "eval_set" column from "orders.csv"
df_ords = df_ords.drop(columns = ['eval_set'])
```

### 3-1-2. Finding the missing values in a column

```python
# using the value_counts() function on the "days_since_prior_order" columns in "df_ords" dataframe to see the missing (NaN) value
df_ords['days_since_prior_order'].value_counts(dropna = False)
```

```python
# Returning the index of the rows
df_ords.index
```

```python
# Returning the count of each nique row in a table
df_ords.value_counts()
```

### 3-1-3. Renaming a column in a dataframe

```python
# Renaming 'order_dow' column into 'orders_day_of_week'.
# The argument "inplace = true", means that we overwrote the original column name with this new column name rather than creating a new copy.
df_ords.rename(columns ={'order_dow' : 'orders_day_of_week'}, inplace = True)
df_ords
```

```python
# Descriptive Statistics for "df_ords"
df_ords.describe()
```

### 3-1-4. Changing a variable’s data type

```python
#changing the data type of the "order_id" column from a number to a string (which the describe() function will ignore when calculating descriptive statistics).
df_ords['order_id'] = df_ords['order_id'].astype('str')
df_ords['order_id'].dtype
```

### 3-1-5. Transposing Data

```python
# Exploring df_dep
df_dep.head()
```

```python
# Transposing "df_dep" via "T" function
df_dep_t = df_dep.T
```

```python
df_dep_t
```

```python
# adding an index to "df_dep_t" dataframe
df_dep_t.reset_index()
```

```python
# Take the first row of "def_dep_t" for the header
new_header = df_dep_t.iloc[0]
new_header
```

```python
# Take the data under the header row for a new df
df_dep_t_new = df_dep_t[1:]
```

```python
df_dep_t_new
```

```python
# Set the header row as the df header
df_dep_t_new.columns = new_header
df_dep_t_new
```

### 3-2-3 Question #4. Finding the bussiest hour for placing order

```python
# First solution would be the mode() function which shows 10:00 AM as the busiest hour for placing order
df_ords['order_hour_of_day'].mode()
```

```python
# Second solution is using the "value_counts()" function.The reults are sorted, by default. The same answer here: 10:00 AM
df_ords['order_hour_of_day'].value_counts()
```
